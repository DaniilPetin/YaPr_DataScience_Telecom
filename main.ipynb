### Названия столбцов

В соответствии с общепринятыми правилами, названия всех столбцов необходимо привести к "змеиному" регистру. В исходнах данных названия содержат заглавные буквы и пробелы (или их неоправданное отсутствие).

```python
def col_rename(df, df_name):
    
    print()
    print(f'\033[1mПереименованные столбцы датасета {df_name}:\033[0m')
    print()

    def camel_to_snake(name):
        name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', str(name))
        name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
        name = name.lower()
        name = re.sub('_+', '_', name)
        if name.startswith('_'):
            name = name[1:]
        return name
    
    df.columns = [camel_to_snake(col) for col in df.columns]
    print(df.columns)
```

```python
col_rename(contract_new, 'contract_new')
col_rename(personal_new, 'personal_new')
col_rename(internet_new, 'internet_new')
col_rename(phone_new, 'phone_new')
```

### Добавление признаков в contract_new

В связи с тем, что данные о конкретных датах начала и окончания действия договора не несут непосредственной ценности для предсказания ухода абонентов в будущем, из них можно получить более полезные и наглядные признаки. 

- Признак `is_left` - разорвал ли абонент договор с оператором на момент сбора информации;
- Признак `duration_contract_days` - продолжительности действия договора в днях.

Таким образом, признак `is_left` будет целевым при решении задачи бинарной классификации для предсказания разрыва договора абонентом оператора связи.
Лишние признаки `begin_date` и `end_date` будут удалены из рассмотрения для более корректного обучения моделей.

```python
contract_new['is_left'] = 0
contract_new.loc[contract_new['end_date'] != '2020-02-01', 'is_left'] = 1

contract_new['is_left'] = contract_new['is_left'].astype('int')
```

```python
contract_new['duration_contract_days'] = (
    (contract_new['end_date'] - contract_new['begin_date'])
        .dt.components.days)
```

```python
contract_new = contract_new.drop(columns=['begin_date', 'end_date'])
contract_new.head()
```

### Подготовка функций

```python
def obj_plotting(df, column):
    print(f"Число и процентное соотношение значений признака '{column}'")
    
    colors = ['#FFD1DC', '#E0F7FA', '#D2F2E3', 
                     '#E1D5E7', '#F5D5CB', '#FFE6E6']
    (
        df.pivot_table(index=column, values='customer_id', aggfunc='count')
        .plot(kind='pie', y='customer_id', figsize=(7, 7), autopct='%1.0f%%', 
              legend=False, colors=colors)
    )
    return df[column].value_counts()
```

```python
def num_plotting(df, column, target):
    print(f"Распределение и боксплот признака '{column}' в зависимости от целевого '{target}':")
    
    display(df[column].describe())
    
    f, axes = plt.subplots(1, 2, figsize=(15, 6))
    axes[0].set_ylabel('количество')
    sns.histplot(df, bins=20, ax=axes[0], hue=target, x=column)
    
    sns.boxplot(data=df, ax=axes[1], y=column)
    axes[1].set_ylabel(column)
    plt.show()
```

```python
def hist_plotting(df, column, target):
    
    plot_data = df[[column, target]].dropna(subset=[column])
    
    print(f'Распределение признака "{column}":')
    
    f, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))
    
    # абсолютные значения
    sns.histplot( plot_data, bins=20, x=column, hue=target, multiple='stack', ax=ax1)
    ax1.set_ylabel('Количество')
    ax1.set_title(f'Распределение признака "{column}" по {target} (абсолютные значения)')
    
    # процентное соотношение
    sns.histplot( plot_data, bins=20, x=column, hue=target, multiple='fill', ax=ax2, stat='probability')
    ax2.set_ylabel('Доля')
    ax2.set_title(f'Распределение признака "{column}" по {target} (процентное соотношение)')
    ax2.set_ylim(0, 1)
    
    for patch in ax2.patches:
        try:
            height = patch.get_height()
            x_center = patch.get_x() + patch.get_width()/2.
            y_center = patch.get_y() + height/2
            
            if (height > 0 and np.isfinite(height) and np.isfinite(x_center) and np.isfinite(y_center)):
                
                ax2.text(x_center, y_center,
                    f'{height:.1%}', ha="center", va="center", fontsize=10, color='black')
                        
        except (ValueError, TypeError):
            continue
    
    plt.tight_layout()
    plt.show()
```

## 5. Подготовка данных

### Разделение данных на выборки

```python
RANDOM_STATE = 250825
TEST_SIZE = 0.25
```

```python
features = ['multiple_lines', 'streaming_movies', 'dependents', 'partner', 'senior_citizen', 
    'duration_contract_days', 'total_charges', 'monthly_charges', 'payment_method', 
    'paperless_billing', 'type']
```

```python
X = df_full[features] # входные признаки
y = df_full['is_left'] # целевой признак

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    random_state = RANDOM_STATE,
    test_size=TEST_SIZE
)
```

### Кодирование и масштабирование признаков

Целевой признак `is_left` создан таким образом, чтобы не потребовалась дополнительная кодировка. Убедимся в этом, прежде чем переходить к работе с остальными признаками

```python
df_full['is_left'].unique()
```

```python
cat_col_names = ['multiple_lines', 'streaming_movies', 'dependents', 'partner', 'senior_citizen',
                 'payment_method', 'paperless_billing', 'type']

num_col_names = ['duration_contract_days', 'total_charges', 'monthly_charges']
```

```python
# кодирование категориальных признаков
encoder = OneHotEncoder(drop='first', sparse_output=False)

# обучение и трансформация категориальных признаков из тренировочной выборки 
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])

# преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names])

# список новых столбцов
encoder_col_names = encoder.get_feature_names_out() 

df_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

display(df_ohe.head())
```

```python
# масштабирование количественных признаков
scaler = StandardScaler()

# обучение и трансформация количественных признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])

# преобразование количественных признаков в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)
```

```python
# объединяем все преобразованные данные в тренировочный и тестовый датафреймы
X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
print('Тренировочная выборка:')
display(X_train.head())
print()
print('Тестовая выборка:')
display(X_test.head())
```

### Вывод по шагу 5

- Подготовленные данные содержат 8 категориальных столбцов и 3 числовых;
- Данные были разделены на обучающие и тестовые в соотношении 3:1;
- Категориальные признаки были закодированы при помощи OneHotEncoder;
- Числовые признаки были отмасштабированы при помощи StandardScaler;
- Целевой прищнак `is_left` не требует кодирование. 'is_left' = 1 соответствует ушедшим пользователям.
- Кодирование и масштабирование были выполнены после разделения данных на выборки, чтобы избежать утечки данных.


Данные готовы к обучению моделей машинного обучения

## 6. Обучение моделей машинного обучения

### LogisticRegression

```python
# инициализация модели логистической регрессии
model_lr = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')
```

```python
# предсказания модели
cv_scores_lr = cross_val_score(model_lr, X_train, y_train, 
                           scoring='roc_auc', cv=5)
model_lr.fit(X_train, y_train)
```

```python
# расчет значения RMSE для линейной регрессии
roc_auc_lr = round(np.abs(np.mean(cv_scores_lr)), 2)
print('Среднее значение ROC-AUC Логистической регрессии =', roc_auc_lr)
```

### KNeighborsClassifier

```python
param_grid_knc = {
    'n_neighbors': range(2, 50, 2),
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
```

```python
grid_search_knc = GridSearchCV(
    estimator = KNeighborsClassifier(),
    param_grid = param_grid_knc,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
)

grid_search_knc.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', grid_search_knc.best_estimator_)
print ('Метрика ROC-AUC лучшей модели на кросс-валидации:', np.abs(grid_search_knc.best_score_))
```

```python
model_knc = grid_search_knc.best_estimator_
roc_auc_knc = np.abs(grid_search_knc.best_score_)
```

### DecisionTreeClassifier

```python
param_grid_dtc = {
    'max_depth': [None] + list(range(3, 20)),
    'max_features': [None, 'sqrt', 'log2', 0.3, 0.5, 0.7],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6, 8]
}
```

```python
grid_search_dtc = GridSearchCV(
    estimator = DecisionTreeClassifier(random_state=RANDOM_STATE, class_weight='balanced'),
    param_grid = param_grid_dtc,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

grid_search_dtc.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', grid_search_dtc.best_estimator_)
print ('Метрика ROC-AUC лучшей модели на кросс-валидации:', np.abs(grid_search_dtc.best_score_))
```

```python
model_dtc = grid_search_dtc.best_estimator_
roc_auc_dtc = np.abs(grid_search_dtc.best_score_)
```

### CatBoostClassifier

```python
param_grid_catboost = {
    'iterations': [300, 500, 700],
    'depth': [4, 6, 8],
    'learning_rate': [0.03, 0.05, 0.1],
    'l2_leaf_reg': [3, 5, 7],
    'random_strength': [1, 3, 5],
    'min_data_in_leaf': [1, 5, 10],
}
```

```python
randomized_search_catboost = RandomizedSearchCV(
    estimator=CatBoostClassifier(random_state=RANDOM_STATE, auto_class_weights='Balanced'),
    param_distributions=param_grid_catboost,
    cv=5,
    n_iter=10,
    scoring='roc_auc',
    n_jobs=-1,
    random_state = RANDOM_STATE
)

randomized_search_catboost.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_catboost.best_estimator_)
print ('Метрика ROC-AUC лучшей модели на кросс-валидации:', np.abs(randomized_search_catboost.best_score_))
```

```python
model_catboost = randomized_search_catboost.best_estimator_
roc_auc_catboost = np.abs(randomized_search_catboost.best_score_)
```
